```python
import tkinter as tk
import tkmacosx as tkm
import numpy as np

calculation = ""

#Input
def add_to_calculation(symbol):
    global calculation
    if calculation == "0":
        clear()
    calculation += str(symbol)
    text_result.delete(1.0, "end")
    text_result.insert(1.0, calculation)

#Output
def evaluate_calculation():
    global calculation
    if calculation == "":
        clear()
    else :
        try:
            calculation = str(eval(calculation))
            text_result.delete(1.0, "end")
            calculation = str(float("%.9f" % float(calculation)))
            if calculation[-2:] == ".0":
                calculation = calculation[:-2]
            text_result.insert(1.0, calculation)
        except:
            calculation = ""
            text_result.delete(1.0, "end")
            text_result.insert(1.0, "Error")

#Clearing Input
def clear():
    global calculation
    if calculation != "":
        calculation = ""
    text_result.delete(1.0, "end")
    text_result.insert(1.0, "0")
    
def backspace():
    global calculation
    if len(calculation)<2:
        clear()
    else :
        calculation = calculation[:-1]
        text_result.delete(1.0, "end")
        text_result.insert(1.0, calculation)

#Calculator
root = tk.Tk()
root.title("Calculator")
root.configure(bg="#000000")

#Calculation Display
text_result = tk.Text(root, height=3, width=37, bg="#000000", fg="#FFFFFF", highlightthickness=1, font=("Cambria Math", 20))
text_result.grid(columnspan=6, pady=10)
text_result.insert(1.0, "0")

#Exit Hint Display
exit_text = tk.Text(root, height=1, width=37, bg="#000000", fg="#545454", tabs=4, highlightthickness=0, font=("Cambria Math", 10))
exit_text.grid(row=8, column=2, columnspan=6, pady=3)
exit_text.insert(1.0, "Press 'Esc' to exit the calculator")

#Exit Key Bind
root.bind('<Escape>', lambda x: root.destroy)

#Numbers
btn_1 = tkm.Button(root, text="1", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(1), font=("Cambria Math", 15))
btn_1.grid(row=2, column=1, padx=1, pady=1)
btn_2 = tkm.Button(root, text="2", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(2), font=("Cambria Math", 15))
btn_2.grid(row=2, column=2, padx=1, pady=1)
btn_3 = tkm.Button(root, text="3", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(3), font=("Cambria Math", 15))
btn_3.grid(row=2, column=3, padx=1, pady=1)
btn_4 = tkm.Button(root, text="4", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(4), font=("Cambria Math", 15))
btn_4.grid(row=3, column=1, padx=1, pady=1)
btn_5 = tkm.Button(root, text="5", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(5), font=("Cambria Math", 15))
btn_5.grid(row=3, column=2, padx=1, pady=1)
btn_6 = tkm.Button(root, text="6", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(6), font=("Cambria Math", 15))
btn_6.grid(row=3, column=3, padx=1, pady=1)
btn_7 = tkm.Button(root, text="7", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(7), font=("Cambria Math", 15))
btn_7.grid(row=4, column=1, padx=1, pady=1)
btn_8 = tkm.Button(root, text="8", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(8), font=("Cambria Math", 15))
btn_8.grid(row=4, column=2, padx=1, pady=1)
btn_9 = tkm.Button(root, text="9", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(9), font=("Cambria Math", 15))
btn_9.grid(row=4, column=3, padx=1, pady=1)
btn_0 = tkm.Button(root, text="0", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(0), font=("Cambria Math", 15))
btn_0.grid(row=5, column=2, padx=1, pady=1)

#Equations
btn_plus = tkm.Button(root, text="+", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("+"), font=("Cambria Math", 15))
btn_plus.grid(row=3, column=4, padx=1, pady=1)
btn_minus = tkm.Button(root, text="-", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("-"), font=("Cambria Math", 15))
btn_minus.grid(row=3, column=5, padx=1, pady=1)
btn_multiply = tkm.Button(root, text="×", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("*"), font=("Cambria Math", 15))
btn_multiply.grid(row=4, column=4, padx=1, pady=1)
btn_divide = tkm.Button(root, text="÷", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("/"), font=("Cambria Math", 15))
btn_divide.grid(row=4, column=5, padx=1, pady=1)
btn_exponentiation = tkm.Button(root, text="^", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("**"), font=("Cambria Math", 15))
btn_exponentiation.grid(row=5, column=4, padx=1, pady=1)
btn_natural_logarithm = tkm.Button(root, text="ln", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.log("), font=("Cambria Math", 15))
btn_natural_logarithm.grid(row=6, column=4, padx=1, pady=1)
btn_sqrt = tkm.Button(root, text="√", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.sqrt("), font=("Cambria Math", 15))
btn_sqrt.grid(row=5, column=5, padx=1, pady=1)
btn_cbrt = tkm.Button(root, text="∛", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.cbrt("), font=("Cambria Math", 15))
btn_cbrt.grid(row=6, column=5, padx=1, pady=1)

#Miscellanious
btn_dot = tkm.Button(root, text=".", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("."), font=("Cambria Math", 15))
btn_dot.grid(row=6, column=1, padx=1, pady=1)
btn_opening_bracket = tkm.Button(root, text="(", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("("), font=("Cambria Math", 15))
btn_opening_bracket.grid(row=5, column=1, padx=1, pady=1)
btn_closing_bracket = tkm.Button(root, text=")", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(")"), font=("Cambria Math", 15))
btn_closing_bracket.grid(row=5, column=3, padx=1, pady=1)
btn_pi = tkm.Button(root, text="π", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.pi"), font=("Cambria Math", 15))
btn_pi.grid(row=6, column=2, padx=1, pady=1)
btn_euler = tkm.Button(root, text="e", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.e"), font=("Cambria Math", 15))
btn_euler.grid(row=6, column=3, padx=1, pady=1)
btn_times_ten_to_the_power_of = tkm.Button(root, text="×10^", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("*10**("), font=("Cambria Math", 15))
btn_times_ten_to_the_power_of.grid(row=7, column=1, padx=1, pady=1)
btn_one_divided_by = tkm.Button(root, text="1/x", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("1/("), font=("Cambria Math", 15))
btn_one_divided_by.grid(row=7, column=2, padx=1, pady=1)
btn_negative = tkm.Button(root, text="(-)", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("(-"), font=("Cambria Math", 15))
btn_negative.grid(row=7, column=3, padx=1, pady=1)

#Different Functions
btn_equal = tkm.Button(root, text="=", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, width=200, command=evaluate_calculation, font=("Cambria Math", 15))
btn_equal.grid(row=7, column=4, columnspan=2, padx=1, pady=1)
btn_clear = tkm.Button(root, text="AC", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, command=clear, font=("Cambria Math", 15))
btn_clear.grid(row=2, column=5, padx=1, pady=1)
btn_delete = tkm.Button(root, text="DEL", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, command=backspace, font=("Cambria Math", 15))
btn_delete.grid(row=2, column=4, padx=1, pady=1)

#Running Calculator
root.mainloop()
```
