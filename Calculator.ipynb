```python
import tkinter as tk
import numpy as np

calculation = ""

#Input
def add_to_calculation(symbol):
    global calculation
    calculation += str(symbol)
    text_result.delete(1.0, "end")
    text_result.insert(1.0, calculation)

#Output
def evaluate_calculation():
    global calculation
    try:
        calculation = str(eval(calculation))
        text_result.delete(1.0, "end")
        if "nan" in calculation:
            clear_calculation()
            text_result.insert(1.0, "Math Error")
        else :
            text_result.insert(1.0, calculation)
    except:
        if "/0" or "/(0" in calculation:
            clear_calculation()
            text_result.insert(1.0, "Undefined")
        else :
            clear_calculation()
            text_result.insert(1.0, "Syntax Error")

#Clearing Input
def clear_calculation():
    global calculation
    calculation = ""
    text_result.delete(1.0, "end")
    
def backspace():
    global calculation
    calculation = calculation[:-1]
    text_result.delete(1.0, "end")
    text_result.insert(1.0, calculation)

#Calculator
root = tk.Tk()
root.title("Calculator")
root.geometry("433x275")

#Calculation Display
text_result = tk.Text(root, height=3, width=32, font=("Cambria Math", 20))
text_result.grid(columnspan=6)

#Numbers
btn_1 = tk.Button(root, text="1", command=lambda: add_to_calculation(1), width=5, font=("Cambria Math", 15))
btn_1.grid(row=2, column=1)
btn_2 = tk.Button(root, text="2", command=lambda: add_to_calculation(2), width=5, font=("Cambria Math", 15))
btn_2.grid(row=2, column=2)
btn_3 = tk.Button(root, text="3", command=lambda: add_to_calculation(3), width=5, font=("Cambria Math", 15))
btn_3.grid(row=2, column=3)
btn_4 = tk.Button(root, text="4", command=lambda: add_to_calculation(4), width=5, font=("Cambria Math", 15))
btn_4.grid(row=3, column=1)
btn_5 = tk.Button(root, text="5", command=lambda: add_to_calculation(5), width=5, font=("Cambria Math", 15))
btn_5.grid(row=3, column=2)
btn_6 = tk.Button(root, text="6", command=lambda: add_to_calculation(6), width=5, font=("Cambria Math", 15))
btn_6.grid(row=3, column=3)
btn_7 = tk.Button(root, text="7", command=lambda: add_to_calculation(7), width=5, font=("Cambria Math", 15))
btn_7.grid(row=4, column=1)
btn_8 = tk.Button(root, text="8", command=lambda: add_to_calculation(8), width=5, font=("Cambria Math", 15))
btn_8.grid(row=4, column=2)
btn_9 = tk.Button(root, text="9", command=lambda: add_to_calculation(9), width=5, font=("Cambria Math", 15))
btn_9.grid(row=4, column=3)
btn_0 = tk.Button(root, text="0", command=lambda: add_to_calculation(0), width=5, font=("Cambria Math", 15))
btn_0.grid(row=5, column=2)

#Equations
btn_plus = tk.Button(root, text="+", command=lambda: add_to_calculation("+"), width=5, font=("Cambria Math", 15))
btn_plus.grid(row=3, column=4)
btn_minus = tk.Button(root, text="-", command=lambda: add_to_calculation("-"), width=5, font=("Cambria Math", 15))
btn_minus.grid(row=3, column=5)
btn_multiply = tk.Button(root, text="×", command=lambda: add_to_calculation("*"), width=5, font=("Cambria Math", 15))
btn_multiply.grid(row=4, column=4)
btn_divide = tk.Button(root, text="÷", command=lambda: add_to_calculation("/"), width=5, font=("Cambria Math", 15))
btn_divide.grid(row=4, column=5)
btn_exponentiation = tk.Button(root, text="^", command=lambda: add_to_calculation("**"), width=5, font=("Cambria Math", 15))
btn_exponentiation.grid(row=5, column=4)
btn_modulo = tk.Button(root, text="mod", command=lambda: add_to_calculation("%"), width=5, font=("Cambria Math", 15))
btn_modulo.grid(row=6, column=4)
btn_sqrt = tk.Button(root, text="√", command=lambda: add_to_calculation("np.sqrt("), width=5, font=("Cambria Math", 15))
btn_sqrt.grid(row=5, column=5)
btn_cbrt = tk.Button(root, text="∛", command=lambda: add_to_calculation("np.cbrt("), width=5, font=("Cambria Math", 15))
btn_cbrt.grid(row=6, column=5)
btn_natural_logarithm = tk.Button(root, text="ln", command=lambda: add_to_calculation("np.log("), width=5, font=("Cambria Math", 15))
btn_natural_logarithm.grid(row=7, column=3)

#Miscellanious
btn_dot = tk.Button(root, text=".", command=lambda: add_to_calculation("."), width=5, font=("Cambria Math", 15))
btn_dot.grid(row=6, column=1)
btn_opening_bracket = tk.Button(root, text="(", command=lambda: add_to_calculation("("), width=5, font=("Cambria Math", 15))
btn_opening_bracket.grid(row=5, column=1)
btn_closing_bracket = tk.Button(root, text=")", command=lambda: add_to_calculation(")"), width=5, font=("Cambria Math", 15))
btn_closing_bracket.grid(row=5, column=3)
btn_pi = tk.Button(root, text="π", command=lambda: add_to_calculation("np.pi"), width=5, font=("Cambria Math", 15))
btn_pi.grid(row=6, column=2)
btn_euler = tk.Button(root, text="e", command=lambda: add_to_calculation("np.e"), width=5, font=("Cambria Math", 15))
btn_euler.grid(row=6, column=3)
btn_times_ten_to_the_power_of = tk.Button(root, text="×10^", command=lambda: add_to_calculation("*10**("), width=5, font=("Cambria Math", 15))
btn_times_ten_to_the_power_of.grid(row=7, column=1)
btn_one_divided_by = tk.Button(root, text="1/x", command=lambda: add_to_calculation("1/("), width=5, font=("Cambria Math", 15))
btn_one_divided_by.grid(row=7, column=2)

#Different Functions
btn_equal = tk.Button(root, text="=", command=evaluate_calculation, width=14, font=("Cambria Math", 15))
btn_equal.grid(row=7, column=4, columnspan=2)
btn_clear = tk.Button(root, text="AC", command=clear_calculation, width=5, font=("Cambria Math", 15))
btn_clear.grid(row=2, column=5)
btn_delete = tk.Button(root, text="DEL", command=backspace, width=5, font=("Cambria Math", 15))
btn_delete.grid(row=2, column=4)
root.mainloop()
```
