```python
import tkinter as tk
import numpy as np

calculation = ""

#Input
def add_to_calculation(symbol):
    global calculation
    if calculation == "0":
        clear()
    calculation += str(symbol)
    text_result.delete(1.0, "end")
    text_result.insert(1.0, calculation)

#Output
def evaluate_calculation():
    global calculation
    if calculation == "":
        clear()
    else :
        try:
            calculation = str(eval(calculation))
            text_result.delete(1.0, "end")
            calculation = str(float("%.9f" % float(calculation)))
            if calculation[-2:] == ".0":
                calculation = calculation[:-2]
            text_result.insert(1.0, calculation)
        except:
            calculation = ""
            text_result.delete(1.0, "end")
            text_result.insert(1.0, "Error")

#Clearing Input
def clear():
    global calculation
    if calculation != "":
        calculation = ""
    text_result.delete(1.0, "end")
    text_result.insert(1.0, "0")
    
def backspace():
    global calculation
    if len(calculation)<2:
        clear()
    else :
        calculation = calculation[:-1]
        text_result.delete(1.0, "end")
        text_result.insert(1.0, calculation)

#Calculator
root = tk.Tk()
root.title("Calculator")
root.configure(bg="#000000")

#Calculation Display
text_result = tk.Text(root, height=2, width=20, bg="#000000", fg="#FFFFFF", highlightthickness=1, font=("Arial", 20))
text_result.grid(columnspan=6, pady=10)
text_result.insert(1.0, "0")

#Exit Hint Display
exit_text = tk.Label(root, bg="#000000", fg="#545454", highlightthickness=0, font=("Arial", 10), text="Press 'Esc' to exit the calculator")
exit_text.grid(row=8, column=2, columnspan=3, pady=3)

#Exit Key Bind
root.bind("<Escape>", lambda x: root.destroy())

#Numbers
btn_1 = tk.Button(root, text="1", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(1), font=("Arial", 15), width=5)
btn_1.grid(row=2, column=1, padx=1, pady=1)
btn_2 = tk.Button(root, text="2", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(2), font=("Arial", 15), width=5)
btn_2.grid(row=2, column=2, padx=1, pady=1)
btn_3 = tk.Button(root, text="3", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(3), font=("Arial", 15), width=5)
btn_3.grid(row=2, column=3, padx=1, pady=1)
btn_4 = tk.Button(root, text="4", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(4), font=("Arial", 15), width=5)
btn_4.grid(row=3, column=1, padx=1, pady=1)
btn_5 = tk.Button(root, text="5", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(5), font=("Arial", 15), width=5)
btn_5.grid(row=3, column=2, padx=1, pady=1)
btn_6 = tk.Button(root, text="6", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(6), font=("Arial", 15), width=5)
btn_6.grid(row=3, column=3, padx=1, pady=1)
btn_7 = tk.Button(root, text="7", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(7), font=("Arial", 15), width=5)
btn_7.grid(row=4, column=1, padx=1, pady=1)
btn_8 = tk.Button(root, text="8", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(8), font=("Arial", 15), width=5)
btn_8.grid(row=4, column=2, padx=1, pady=1)
btn_9 = tk.Button(root, text="9", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(9), font=("Arial", 15), width=5)
btn_9.grid(row=4, column=3, padx=1, pady=1)
btn_0 = tk.Button(root, text="0", bg="#000000", fg="#FFFFFF", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(0), font=("Arial", 15), width=5)
btn_0.grid(row=5, column=2, padx=1, pady=1)

#Equations
btn_plus = tk.Button(root, text="+", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("+"), font=("Arial", 15), width=5)
btn_plus.grid(row=3, column=4, padx=1, pady=1)
btn_minus = tk.Button(root, text="-", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("-"), font=("Arial", 15), width=5)
btn_minus.grid(row=3, column=5, padx=1, pady=1)
btn_multiply = tk.Button(root, text="×", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("*"), font=("Arial", 15), width=5)
btn_multiply.grid(row=4, column=4, padx=1, pady=1)
btn_divide = tk.Button(root, text="÷", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("/"), font=("Arial", 15), width=5)
btn_divide.grid(row=4, column=5, padx=1, pady=1)
btn_exponentiation = tk.Button(root, text="^", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("**"), font=("Arial", 15), width=5)
btn_exponentiation.grid(row=5, column=4, padx=1, pady=1)
btn_natural_logarithm = tk.Button(root, text="ln", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.log("), font=("Arial", 15), width=5)
btn_natural_logarithm.grid(row=6, column=4, padx=1, pady=1)
btn_sqrt = tk.Button(root, text="√", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.sqrt("), font=("Arial", 15), width=5)
btn_sqrt.grid(row=5, column=5, padx=1, pady=1)
btn_cbrt = tk.Button(root, text="∛", bg="#000000", fg="#2CC943", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.cbrt("), font=("Arial", 15), width=5)
btn_cbrt.grid(row=6, column=5, padx=1, pady=1)

#Miscellanious
btn_dot = tk.Button(root, text=".", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("."), font=("Arial", 15), width=5)
btn_dot.grid(row=6, column=1, padx=1, pady=1)
btn_opening_bracket = tk.Button(root, text="(", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("("), font=("Arial", 15), width=5)
btn_opening_bracket.grid(row=5, column=1, padx=1, pady=1)
btn_closing_bracket = tk.Button(root, text=")", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation(")"), font=("Arial", 15), width=5)
btn_closing_bracket.grid(row=5, column=3, padx=1, pady=1)
btn_pi = tk.Button(root, text="π", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.pi"), font=("Arial", 15), width=5)
btn_pi.grid(row=6, column=2, padx=1, pady=1)
btn_euler = tk.Button(root, text="e", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("np.e"), font=("Arial", 15), width=5)
btn_euler.grid(row=6, column=3, padx=1, pady=1)
btn_times_ten_to_the_power_of = tk.Button(root, text="×10^", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("*10**("), font=("Arial", 15), width=5)
btn_times_ten_to_the_power_of.grid(row=7, column=1, padx=1, pady=1)
btn_one_divided_by = tk.Button(root, text="1/x", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("1/("), font=("Arial", 15), width=5)
btn_one_divided_by.grid(row=7, column=2, padx=1, pady=1)
btn_negative = tk.Button(root, text="(-)", bg="#000000", fg="#F7B82F", activebackground='#141414', takefocus=0, command=lambda: add_to_calculation("(-"), font=("Arial", 15), width=5)
btn_negative.grid(row=7, column=3, padx=1, pady=1)

#Different Functions
btn_equal = tk.Button(root, text="=", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, width=11, command=evaluate_calculation, font=("Arial", 15))
btn_equal.grid(row=7, column=4, columnspan=2, padx=1, pady=1)
btn_clear = tk.Button(root, text="AC", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, width=5, command=clear, font=("Arial", 15))
btn_clear.grid(row=2, column=5, padx=1, pady=1)
btn_delete = tk.Button(root, text="DEL", bg="#000000", fg="#F85B54", activebackground='#141414', takefocus=0, width=5, command=backspace, font=("Arial", 15))
btn_delete.grid(row=2, column=4, padx=1, pady=1)

#Running Calculator
root.mainloop()
```
